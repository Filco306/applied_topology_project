```python
def arccosdist(vect1, vect2):
    if (vect1 == vect2).all():
        return 0
    return math.degrees(np.arccos(1 - dist.cosine(vect1, vect2)))
```

```python
def euclid_dist(vect1, vect2):
    if (vect1 == vect2).all():
        return 0
    return np.linalg.norm(vect1 - vect2)
```
