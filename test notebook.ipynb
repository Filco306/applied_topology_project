```python
from graph import *
import spacy
import scipy.spatial.distance as dist
```

```python
model = spacy.load('en_vectors_glove_md')
```

```python
x = next(iter(model.vocab))
```

```python
def most_similar(word, model, threshold):
    vect = model(word).vector
    words = [x.text for x in model.vocab if dist.cosine(x.vector, vect) <= threshold]
    return words
```

```python
most_similar('bank', model, threshold=0.1)
```
