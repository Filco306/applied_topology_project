```python
def most_similar(word, model, threshold):
    vect = model(word).vector
    words = [x.text for x in model.vocab if dist.cosine(x.vector, vect) <= threshold]
    return words
```

```python
def most_similar_arccos(word, model, threshold):
    vect = model(word).vector
    words = [x.text for x in model.vocab if math.degrees(np.arccos(1 - dist.cosine(x.vector, vect))) <= threshold]
    return words
```

```python
data = list(set([x.lower() for x in data]))
```

```python
def get_distance_matrix():
#     USES ARCCOS-DIST
    df = pd.read_csv('/home/tadas/words.txt')
    df = pd.DataFrame(data=data)

    distance_matrix = []

    for i, word in enumerate(df[0]):
        # print(i)
        vector = model(word).vector
        row = []
        for word_2 in df[0]:
            vector_2 = model(word_2).vector
            if word == word_2:
                row.append(0)
                continue
            vector = vector/np.linalg.norm(vector)
            vector_2 = vector_2/np.linalg.norm(vector_2)
            distance = math.degrees(np.arccos(1 - dist.cosine(vector, vector_2)))
            row.append(distance)
        distance_matrix.append(row)

    distance_matrix = np.array(distance_matrix)
    return df, distance_matrix
```

```python
df, matrix = get_distance_matrix() 
```

```python
np.mean(matrix)
```

```python
np.std(matrix)
```

```python
dg = DataGraph(matrix, 56, 300)
```

```python
dg.cluster(report_homology=True)
```

```python
new_clusters = []                               
for cluster in dg.clusters:  
    cluster, hom = cluster                                                                  
    new_cluster = [df[0][x] for x in cluster]
    new_clusters.append((new_cluster, hom[:10]))
```
